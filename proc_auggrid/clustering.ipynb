```python
import pandas as pd
import numpy as np

df = pd.read_csv('AusGrid_preprocess.csv', header=[0,1], index_col=0)
df = df.set_index(pd.to_datetime(df.index))
df.columns = df.columns.set_levels(df.columns.levels[0].astype('int64'), level=0)
df.head()
```

```python
customers = sorted(df.columns.levels[0])
gg = pd.DataFrame()
for c in customers:
    gg = pd.concat([gg, df[c].GG.rename(c).to_frame().T])
gg
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

num_cluster = 30
inertias = []
for i in range(1,num_cluster):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(gg)
    inertias.append(kmeans.inertia_)

plt.plot(range(1,num_cluster), inertias, marker='o')
plt.title('Elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()

```

```python
NC = 10
kmeans = KMeans(n_clusters=NC)
kmeans.fit(gg)

print(kmeans.labels_)
```

```python
plt.figure(figsize=(16,8))

for i in range(len(customers)):
    plt.subplot(2, 5, kmeans.labels_[i]+1)
    d = gg.iloc[[i]].T
    plt.plot(d.resample('1D').max())
    plt.ylim(0, 9000)

    
```
